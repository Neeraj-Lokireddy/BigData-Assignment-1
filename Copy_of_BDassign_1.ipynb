```python
import plotly.express as px
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Neeraj').getOrCreate()
```

```python
state_codes = {
        'Alabama': 'AL', 'Alaska': 'AK', 'American Samoa': 'AS', 'Arizona': 'AZ', 'Arkansas': 'AR', 'California': 'CA',
        'Colorado': 'CO', 'Connecticut': 'CT', 'Delaware': 'DE', 'Diamond Princess':'DP', 'District of Columbia':'DC', 'Florida': 'FL', 'Georgia': 'GA',
        'Grand Princess': 'GP', 'Guam': 'GU', 'Hawaii': 'HI', 'Idaho': 'ID', 'Illinois': 'IL', 'Indiana': 'IN', 'Iowa': 'IA',
        'Kansas': 'KS', 'Kentucky': 'KY', 'Louisiana': 'LA', 'Maine': 'ME', 'Maryland': 'MD',
        'Massachusetts': 'MA', 'Michigan': 'MI', 'Minnesota': 'MN', 'Mississippi': 'MS',
        'Missouri': 'MO', 'Montana': 'MT', 'Nebraska': 'NE', 'Nevada': 'NV', 'New Hampshire': 'NH',
        'New Jersey': 'NJ', 'New Mexico': 'NM', 'New York': 'NY', 'North Carolina': 'NC',
        'North Dakota': 'ND', 'Northern Mariana Islands': 'MP', 'Ohio': 'OH', 'Oklahoma': 'OK', 'Oregon': 'OR', 'Pennsylvania': 'PA',
        'Puerto Rico': 'PR', 'Rhode Island': 'RI', 'South Carolina': 'SC', 'South Dakota': 'SD', 'Tennessee': 'TN',
        'Texas': 'TX', 'Utah': 'UT', 'Vermont': 'VT', 'Virgin Islands': 'VI', 'Virginia': 'VA', 'Washington': 'WA',
        'West Virginia': 'WV', 'Wisconsin': 'WI', 'Wyoming': 'WY'
    }
hi=['']*58
sf = spark.read.csv('01-01-2021.csv', header = True, inferSchema = True)
s_names = sf.select("Province_State")
x=-1
for i in s_names.collect():
  x+=1
  r=(i.asDict()["Province_State"])
  hi[x]=state_codes[r]
```

```python
months=[0,31,28,31,30,31,30,31,31,30,31,30,31]
for mo in range(1,13):
  rang=months[mo]+1
  if mo<10:
    month='0'+str(mo)
  else:
    month=str(mo)


  hello=[0]*58
  for i in range(1,rang):
    if i<10:
      s='0'+str(i)
    else:
      s=str(i)
    path = month+'-'+s+'-2021.csv'
    df = spark.read.csv(path, header = True, inferSchema = True)
    sc = df.select("Case_Fatality_Ratio")
    x=-1
    for i in sc.collect():
      x+=1
      r=(i.asDict()["Case_Fatality_Ratio"])
      if r==None:
        hello[x]+=0
      else:
        hello[x]+=r
  final_list=[]
  for i,j in zip(hi,hello):
    final_list.append([i,j])
  s_l = sorted(final_list, key=lambda x: x[1], reverse= True)
  print('The Top Ten States with highest IFR are:')
  for o in (s_l[:11]):
    print(o[0],'-->',o[1])
  m_dict = {'Province_State': hi,
          'Case_Fatality_Ratio': hello

  }

  fig = px.choropleth(m_dict,
                      locations='Province_State',
                      locationmode='USA-states',
                      scope='usa',
                      color='Case_Fatality_Ratio',
                      color_continuous_scale='rainbow')
  fig.show()
```
